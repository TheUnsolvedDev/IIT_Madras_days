# Question 2: Gradient descent and Ridge Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from tqdm import tqdm
import random
```

```python
data = loadmat('q2_train.mat')
data['train'].shape
```

```python
data_test = loadmat('q2_test.mat')
```

```python
def do_gradient_descent(data,labels,theta,lr,n_points,n_iter):
    cost_hist = []
    for i in tqdm(range(n_iter),ncols=120):
        proj_feature = np.dot(data,theta)
        # avg cost per example (the 2 in 2*n_points doesn't really matter here.
        # But to be consistent with the gradient, I include it)
        loss = proj_feature - labels
        cost = np.sum(loss**2)/(2*n_points)
        cost_hist.append(cost)
        grad = np.dot(data.T,loss)/n_points
        # update
        theta = theta-lr*grad
    return theta, cost_hist
```

```python
def do_stoch_gradient_descent(data,labels,theta,lr,n_points,n_iter,batch_size=100):
    cost_hist = []
    for i in tqdm(range(0,n_iter,batch_size),ncols=120):
        x_mod,y_mod = data[i:i+batch_size,:], labels[i:i+batch_size]
        proj_feature = np.dot(x_mod,theta)
        # avg cost per example (the 2 in 2*n_points doesn't really matter here.
        # But to be consistent with the gradient, I include it)
        loss = proj_feature - y_mod
        cost = np.sum(loss**2)/(2*n_points)
        cost_hist.append(cost)
        grad = np.dot(x_mod.T,loss)/n_points
        # update
        theta = theta-lr*grad
    return theta, cost_hist
```

```python
def do_ridge_regression(data,labels,theta,lr,n_points,n_iter,lamda_=0.1):
    cost_hist = []
    for i in tqdm(range(n_iter),ncols=120):
        proj_feature = np.dot(data,theta)
        # avg cost per example (the 2 in 2*n_points doesn't really matter here.
        # But to be consistent with the gradient, I include it)
        loss = proj_feature - labels
        cost = np.sum(loss**2)/(2*n_points) + lamda_*np.linalg.norm(theta)**2
        cost_hist.append(cost)
        grad = np.dot(data.T,loss)/n_points
        # update
        theta = theta-lr*grad
    return theta, cost_hist
```

```python
m,n = np.shape(data['train'])
numIterations= int(5e4)
alpha = 0.0005
theta = np.random.randn(n)
theta_gd,cost_hist_gd = do_gradient_descent(data['train'], data['label'][:,0], theta, alpha, m, numIterations)
```

```python
m,n = np.shape(data['train'])
numIterations= int(5e4)
alpha = 0.0005
theta = np.random.randn(n)
theta_sgd,cost_hist_sgd = do_stoch_gradient_descent(data['train'], data['label'][:,0], theta, alpha, m, numIterations,100)
```

```python
m,n = np.shape(data['train'])
numIterations= int(5e4)
alpha = 0.0005
theta = np.random.randn(n)
lamda_list = [1,1e-1,1e-2,1e-3]
theta_rr,cost_hist_rr = np.zeros((n,len(lamda_list))),np.zeros((numIterations,len(lamda_list)))
for i in range(len(lamda_list)):
    print('trying lamda:',lamda_list[i])
    theta_rr[:,i],cost_hist_rr[:,i] = do_ridge_regression(data['train'], data['label'][:,0], theta, alpha, m, numIterations,lamda_list[i])
theta_rr.shape
```

```python
plt.figure(figsize=(10,7))
for i in range(len(lamda_list)):
    plt.plot(cost_hist_rr[:,i],label='lamda={_i}'.format(_i=lamda_list[i]))

plt.grid()
_ = plt.xticks(fontsize=14)
_ = plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.xscale('log')
plt.ylabel('Cost',fontsize=16)
plt.xlabel('#iters',fontsize=16)
```

```python
plt.figure(figsize=(10,7))
plt.plot(cost_hist_gd,label='Grad Descent')
plt.plot(cost_hist_sgd,label='Stoch Grad Descent')
plt.plot(cost_hist_rr[:,-1],label='Ridge Regression (best)')
plt.grid()
_ = plt.xticks(fontsize=14)
_ = plt.yticks(fontsize=14)
plt.legend(fontsize=14)
# plt.xscale('log')
plt.ylabel('Cost',fontsize=16)
plt.xlabel('#iters',fontsize=16)
```

```python
pred_label_gd = np.matmul(data_test['test'], theta_gd[:,np.newaxis])
pred_label_sgd = np.matmul(data_test['test'], theta_sgd[:,np.newaxis])
pred_label_rr = np.matmul(data_test['test'], theta_rr[:,-1][:,np.newaxis])
```

```python
mse_gd = (np.square(pred_label_gd - data_test['labels'])).mean(axis=0)
mse_sgd = (np.square(pred_label_sgd - data_test['labels'])).mean(axis=0)
mse_rr = (np.square(pred_label_rr - data_test['labels'])).mean(axis=0)

print('MSE Gradient Descent: ',mse_gd)
print('MSE Stochastic Gradient Descent: ',mse_sgd)
print('MSE Ridge Regression: ',mse_rr)
```
