```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(a,b,c, x):
    if x-b>=0:
        return a*(x-b)+c
   
    return -a*(x-b)+c

def grad(a,b,x):
    if x-b>=0:
        return a
   
    return -a
```

```python
# For case 1
alpha = 2
rate = 0.9

a,b,c = 1,2,3

tolerance = 10**-4
x = 8
xStar = b
count = 1
error = []
prev = 0
lst_x = np.array([x])

while abs(x-xStar)>tolerance and count<100:
    prev = x
    x = x - alpha*grad(a,b,x)
    lst_x = np.append(lst_x, x)
    
    error.append(abs(x-xStar))
    
    alpha*=rate
    count+=1

print(f'The optimum value was found to be(for a =1, b=2 and c=3): {x}')

plt.figure(dpi = 400)
plt.plot(range(len(error)), error)
plt.xlabel('x')
plt.ylabel('error')
plt.title('Plot of the error for a = 1, b = 2 and c = 3')

y = a*abs(lst_x-b*np.ones(lst_x.shape)) + c*np.ones(lst_x.shape)

plt.figure(dpi = 400)
plt.plot(lst_x,y)
plt.xlabel('x_t')
plt.ylabel('f(x_t)')
plt.title('Plot of the function for a = 1, b = 2 and c = 3')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(a,b,c, x):
    return (x**4)+0.5*(x**3)-4*(x**2)

def grad(a,b,x):
    return 4*(x**3)+1.5*(x**2)-8*x

def double_grad(a,b,x):
    return 12*(x**2)+3*(x)-8
```

```python
# For case 1(i)
x_locmin = 1.23909
x_glbmin = -1.61409
tolerance = 10**-4
x = 2
count = 1
error = []
rate = []
prev = x
lst_x = np.array([x])

while abs(x-x_glbmin)>tolerance and count<100:
  
    x = x - (grad(a,b,x)/double_grad(a,b,x))
    lst_x = np.append(lst_x, x)
    error.append(abs(x-x_glbmin))
    
    count+=1

print(f'The optimum value was found to be: {x}')

plt.figure(dpi = 40)
plt.plot(range(len(error)), error)
plt.xlabel('x')
plt.ylabel('error')
plt.title('Plot of the error')
y = (lst_x**4)+0.5*(lst_x**3) - 4*(lst_x**2)

plt.figure(dpi = 40)
plt.plot(lst_x,y)
plt.xlabel('x_t')
plt.ylabel('f(x_t)')
plt.title('Plot of the function')

for i in range(len(error)-1):
  rate.append(error[i]/(error[i+1]**2))

plt.figure(dpi = 40)
plt.plot(range(len(rate)), rate)
plt.xlabel('x')
plt.ylabel('rate')
plt.title('Plot of the rate')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to optimize (example: a simple quadratic function)
def function_to_optimize(x):
    return x**2 + 2*x + 1

# Define the derivative of the function (gradient)
def gradient(x):
    return 2*x + 2

# Perform gradient descent optimization
def gradient_descent(learning_rate=0.1, num_steps=20, start_x=-3):
    x_values = [start_x]
    for _ in range(num_steps):
        gradient_value = gradient(x_values[-1])
        new_x = x_values[-1] - learning_rate * gradient_value
        x_values.append(new_x)
    return np.array(x_values)

# Perform gradient descent
steps = gradient_descent(learning_rate=0.3, num_steps=20)

# Create quiver plot
x = np.linspace(-4, 2, 400)
y = function_to_optimize(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='black', label='Function to Optimize')
plt.scatter(steps, function_to_optimize(steps), color='red', label='Gradient Descent Steps')
plt.quiver(steps[:-1], function_to_optimize(steps[:-1]), steps[1:]-steps[:-1], function_to_optimize(steps[1:])-function_to_optimize(steps[:-1]), scale=1, scale_units='xy', angles='xy', color='blue', label='Gradient Direction')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gradient Descent with Quiver Plot')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rosenbrock(x, y):
    """
    Rosenbrock function: f(x, y) = (a - x)^2 + b * (y - x^2)^2
    """
    a = 1
    b = 100
    return (a - x)**2 + b * (y - x**2)**2

# Generate values for x and y
x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x, y)

# Compute the Rosenbrock function values for the entire grid
Z = rosenbrock(X, Y)

# Plot the Rosenbrock function
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=50, cmap='viridis')  # Contour plot for Rosenbrock function
plt.colorbar(label='Rosenbrock Value')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Rosenbrock Function')
plt.show()
```
